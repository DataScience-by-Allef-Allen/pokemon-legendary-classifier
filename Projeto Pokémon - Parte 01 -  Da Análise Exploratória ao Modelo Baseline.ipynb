## Análise Exploratória de Dados (EDA):



### Estatísticas descritivas individuais

```python
numeric_cols = ['hp', 'attack', 'defense', 'speed', 'special_attack', 'special_defense', 'generation']

means = df[numeric_cols].mean()
stds = df[numeric_cols].std()
max_vals = df[numeric_cols].max()
min_vals = df[numeric_cols].min()

print("\nMédias:")
print(means)
print("\nDesvios Padrão:")
print(stds)
print("\nValores Máximos:")
print(max_vals)
print("\nValores Mínimos:")
print(min_vals)
```

### Matriz de correlação

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df[numeric_cols].corr(method='spearman') #pode alterar para pearson(padrão), spearman ou kendall
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Heatmap de Correlação entre Atributos Numéricos')
plt.show()
```

### Histograma

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

colors = ['blue', 'red', 'green', 'purple', 'orange', 'darkcyan', 'crimson', 'teal']

for i, col in enumerate(numeric_cols):
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, kde=True, bins=30,
                 color=colors[i % len(colors)], edgecolor='black', alpha=1)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.tight_layout()
    plt.show()
```

### Gráfico para analisar as características de pokémons não lendários e lendários

```python
legendary_stats = df.groupby('is_legendary')[numeric_cols].mean()

print("\nMédias dos atributos para Pokémon nÃO Lendários vs Lendários:")
print(legendary_stats)

legendary_stats.plot(kind='bar', figsize=(12, 6))
plt.title('Médias dos Atributos: Não Lendários vs Lendários')
plt.ylabel('Valor Médio')
plt.xlabel('Lendário')
plt.xticks(ticks=[0, 1], labels=['Não Lendário', 'Lendário'], rotation=0)
plt.legend(title='Atributos')
plt.show()
```

## Pré-processamento:

### Engenharia de atributos e codificação de variáveis categóricas

```python
import ast

df['base_stat_total'] = df[['hp', 'attack', 'defense', 'speed', 'special_attack', 'special_defense']].sum(axis=1)
df['num_types'] = df['types'].apply(lambda x: len(ast.literal_eval(x)))
df['num_weaknesses'] = df['weaknesses'].apply(lambda x: len(eval(x)))
df['num_evolutions'] = df['evolutions'].apply(lambda x: len(eval(x)))
df['is_non_evolving'] = df['num_evolutions'].apply(lambda x: 1 if x == 1 else 0)
#df['is_mythical'] = df['is_mythical'].astype(int)

types_of_interest = ['dragon', 'psychic', 'fairy', 'fire', 'electric']
for type_name in types_of_interest:
    df[f'is_{type_name}_type'] = df['types'].apply(lambda x: 1 if type_name in eval(x) else 0)

#drop_cols = ['name', 'types', 'weaknesses', 'evolutions', 'is_mythical']
#df = df.drop(columns=drop_cols)

df = df[['generation','hp', 'attack', 'defense', 'speed', 'special_attack', 'special_defense',
    'base_stat_total', 'num_types', 'num_weaknesses', 'num_evolutions', 'is_non_evolving',
    'is_dragon_type', 'is_psychic_type', 'is_fairy_type', 'is_fire_type', 'is_electric_type', 'is_legendary']]

df.to_csv('pokedex_codificada.csv', index=False)
```

## Divisão , normalização ou padronização, balanceamento dos dados e baseline

### Separar features (X) e alvo (y)

```python
X = df.drop(columns=['is_legendary'])
y = df['is_legendary']
```

```python
print(f"Shape do DataFrame de features: {X.shape}")
display(X)
```

### Divisão treino e teste com estratificação

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

### Normalização com MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
X_train_minmax = minmax_scaler.fit_transform(X_train)
X_test_minmax = minmax_scaler.transform(X_test)
```

### Padronização com StandardScaler

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
X_train_standard = standard_scaler.fit_transform(X_train)
X_test_standard = standard_scaler.transform(X_test)
```

### Recriar DataFrames e salvar como CSV

```python
X_train_df = pd.DataFrame(X_train_minmax, columns=X.columns)
X_test_df = pd.DataFrame(X_test_minmax, columns=X.columns)

train_df = X_train_df.copy()
train_df['is_legendary'] = y_train.values

test_df = X_test_df.copy()
test_df['is_legendary'] = y_test.values

train_df.to_csv('poke_train_minmax.csv', index=False)
test_df.to_csv('poke_test_minmax.csv', index=False)
```

### Balanceamento de vlasses com SMOTE (Undersampling (comentado))

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler


X_train = pd.read_csv('poke_train_minmax.csv').drop(columns='is_legendary')
y_train = pd.read_csv('poke_train_minmax.csv')['is_legendary']

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

df_train_smote = pd.DataFrame(X_train_smote, columns=X_train.columns)
df_train_smote['is_legendary'] = y_train_smote
df_train_smote.to_csv('poke_train_minmax_smote.csv', index=False)

# --- Alternativa: Undersampling (comentado) ---
# undersampler = RandomUnderSampler(random_state=42)
# X_train_under, y_train_under = undersampler.fit_resample(X_train, y_train)

# df_train_under = pd.DataFrame(X_train_under, columns=X_train.columns)
# df_train_under['is_legendary'] = y_train_under
# df_train_under.to_csv('poke_train_minmax_undersampling.csv', index=False)
```

## Baseline de Classificação

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


train_df = pd.read_csv('poke_train_minmax_smote.csv')
test_df = pd.read_csv('poke_test_minmax.csv')

X_train = train_df.drop(columns='is_legendary')
y_train = train_df['is_legendary']

X_test = test_df.drop(columns='is_legendary')
y_test = test_df['is_legendary']

baseline = DummyClassifier(strategy='most_frequent')
baseline.fit(X_train, y_train)

y_pred = baseline.predict(X_test)

print("Acurácia do baseline:", accuracy_score(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred, zero_division=0))


cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Lendário', 'Lendário'], yticklabels=['Não Lendário', 'Lendário'])
plt.xlabel('Classe Predita')
plt.ylabel('Classe Real')
plt.title('Matriz de Confusão - Baseline')
plt.tight_layout()
plt.show()
```
