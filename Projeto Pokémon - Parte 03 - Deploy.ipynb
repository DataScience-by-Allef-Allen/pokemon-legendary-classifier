```python
import gradio as gr
import joblib
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="gradio")



model = joblib.load('best_model_xgb_rand.pkl')

def predict_legendary(generation, hp, attack, defense, speed, special_attack, special_defense,
                    base_stat_total, num_types, num_weaknesses, num_evolutions, is_non_evolving,
                    is_dragon_type, is_psychic_type, is_fairy_type, is_fire_type, is_electric_type):
    input = pd.DataFrame([[generation, hp, attack, defense, speed, special_attack, special_defense,
                             base_stat_total, num_types, num_weaknesses, num_evolutions, is_non_evolving,
                             is_dragon_type, is_psychic_type, is_fairy_type, is_fire_type, is_electric_type]],
                           columns=['generation','hp', 'attack', 'defense', 'speed', 'special_attack', 'special_defense',
                                    'base_stat_total', 'num_types', 'num_weaknesses', 'num_evolutions', 'is_non_evolving',
                                    'is_dragon_type', 'is_psychic_type', 'is_fairy_type', 'is_fire_type', 'is_electric_type'])

    prob = model.predict_proba(input)[0][1]
    classes = 'Lendário' if prob > 0.735 else 'Não Lendário'

    return f"{classes} (Confiança: {prob:.2%})"


interface = gr.Interface(fn=predict_legendary,
                     inputs=[gr.Slider(1, 9, step=1, value=1, label='Generation'),
                             gr.Slider(0, 255, 100, label='HP'),
                             gr.Slider(0, 181, 100, label='Attack'),
                             gr.Slider(0, 230, 100, label='Defense'),
                             gr.Slider(0, 200, 100, label='Speed'),
                             gr.Slider(0, 173, 100, label='Special Attack'),
                             gr.Slider(0, 230, 100, label='Special Defense'),
                             gr.Slider(0, 720, 100, label='Base Stat Total'),
                             gr.Slider(1, 2, step=1, value=1, label='Number of Types'),
                             gr.Slider(0, 10, step=1, value=3, label='Number of Weaknesses'),
                             gr.Slider(1, 3, step=1, value=1, label='Number of Evolutions'),
                             gr.Checkbox(label='Non-Evolving'),
                             gr.Checkbox(label='Dragon Type'),
                             gr.Checkbox(label='Psychic Type'),
                             gr.Checkbox(label='Fairy Type'),
                             gr.Checkbox(label='Fire Type'),
                             gr.Checkbox(label='Electric Type')],
                     outputs=[gr.Textbox()],
                     allow_flagging="never")

interface.launch()
```
